**Program - 1 Census**

```python
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'Age': ['0-4', '5-9', '10-14', '15-19', '20-24', '25-29'],
    'Population': [120000, 150000, 180000, 200000, 250000, 220000]
}

df = pd.DataFrame(data)

plt.bar(df['Age'], df['Population'], color='skyblue')
plt.xlabel('Age Group')
plt.ylabel('Population')
plt.title('Population Distribution by Age Group')
plt.show()
```

**Program - 2 Geospatial**

```python
import geopandas as gpd
import matplotlib.pyplot as plt

data = {
    'City': ['City A', 'City B', 'City C', 'City D'],
    'Country': ['Country 1', 'Country 2', 'Country 1', 'Country 3'],
    'Latitude': [34.0522, 40.7128, 41.8781, 37.7749],
    'Longitude': [-118.2437, -74.0060, -87.6298, -122.4194],
}

geo_df = gpd.GeoDataFrame(data, geometry=gpd.points_from_xy(data['Longitude'], data['Latitude']))

plt.figure(figsize=(10, 6))
geo_df.plot(ax=plt.gca(), marker='o', color='red', markersize=100, label='Cities')
plt.title('Static Geospatial Data Visualization')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

for idx, row in geo_df.iterrows():
    plt.text(row.geometry.x + 0.5, row.geometry.y, row['City'], fontsize=9, color='blue')

plt.legend()
plt.grid(True)
plt.show()
```
