# **Exp - 11 Text visualization using web analytics.**

Text visualization can help in understanding web analytics by transforming textual data into visually interpretable formats like word clouds, bar charts, or heatmaps. It highlights key patterns, trends, and user behavior, making it easier to identify frequently visited pages, popular search terms, or common user feedback. For example, a word cloud of search queries can reveal user interests, helping to optimize content strategies. Visualizing bounce rates or user comments can also uncover areas of improvement for website engagement. By simplifying complex text data, visualization enhances decision-making and supports targeted strategies to improve website performance and user experience.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/Web Analytic_Dataset.csv')
```

```python
metrics = ['Users', 'Pageviews', 'Bounce Rate']

plt.figure(figsize=(12, 6))
for metric in metrics:
    plt.bar(df['Month of the year'], df[metric], label=metric, alpha=0.7)

plt.title('Web Analytics Metrics Over Time')
plt.xlabel('Month of the year')
plt.ylabel('Metrics Value')
plt.legend()
plt.show()
```
